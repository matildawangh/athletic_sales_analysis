```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
data_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
data_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_data = pd.concat([data_2020, data_2021], axis=0)

# Reset the index
combined_data.reset_index(drop=True, inplace=True)
combined_data
```

```python
# Check if any values are null.
has_null_value = combined_data.isnull().any()
has_null_value
```

```python
# Check the data type of each column
data_types = combined_data.dtypes
data_types
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_data['invoice_date'] = pd.to_datetime(combined_data['invoice_date'])
coverted_data = combined_data.dtypes

```

```python
# Confirm that the "invoice_date" data type has been changed.
coverted_data
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".


groupby_data = combined_data.groupby(['region', 'state', 'city'])[['units_sold']].sum()
                            
# Show the top 5 results.
groupby_data_sorted = groupby_data.sort_values(by='units_sold', ascending=False)  
groupby_data_sorted
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_sales = pd.pivot_table(combined_data,values='units_sold', index=['region', 'state', 'city'], aggfunc='sum').reset_index()
# Rename the "units_sold" column to "Total_Products_Sold"
pivot_sales.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# # Show the top 5 results
pivot_sales_top5 = pivot_sales.nlargest(5, 'Total_Products_Sold')
pivot_sales_top5 


```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_total_sales = combined_data.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
grouped_total_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the top 5 results.
grouped_total_sales_top5 = grouped_total_sales.nlargest(5, 'Total Sales')
grouped_total_sales_top5 



```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_table_total_sales = pd.pivot_table(combined_data, index=['region', 'state', 'city'],values='total_sales', aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_total_sales.rename(columns={'total_sales':'Total Sales'}, inplace=True)

# Show the top 5 results.
pivot_table_total_sales_top5= pivot_table_total_sales.nlargest(5,'Total Sales')
pivot_table_total_sales_top5
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales = combined_data[combined_data['product'] == "Women's Athletic Footwear"]
womens_footwear_sales
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Show the top 5 results.

# Filter for Women's Athletic Footwear
womens_footwear_sales = combined_data[combined_data['product'] == "Women's Athletic Footwear"]

# Group by retailer, region, state, and city, then sum the units sold
grouped_womens_footwear_sales = womens_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the "units_sold" column
grouped_womens_footwear_sales.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Get the top 5 results by Womens_Footwear_Units_Sold
top5_womens_footwear_sales = grouped_womens_footwear_sales.nlargest(5, 'Womens_Footwear_Units_Sold')

top5_womens_footwear_sales

```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.


# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

# Show the top 5 results.



# Filter for Women's Athletic Footwear
womens_footwear_sales = combined_data[combined_data['product'] == "Women's Athletic Footwear"]

# Create a pivot table to sum units sold, indexed by retailer, region, state, and city
pivot_womens_footwear_sales = pd.pivot_table(
    womens_footwear_sales, 
    values='units_sold', 
    index=['retailer', 'region', 'state', 'city'], 
    aggfunc='sum'
).rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Sort the pivot table to get the top 5 results by Womens_Footwear_Units_Sold
top5_womens_footwear_sales = pivot_womens_footwear_sales.nlargest(5, 'Womens_Footwear_Units_Sold')

top5_womens_footwear_sales


```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table_by_date = pd.pivot_table(combined_data, 
                                     index='invoice_date', 
                                     values='total_sales', 
                                     aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_by_date.rename(columns={'total_sales': 'Total Sales'}, inplace=True)


# Show the table.
pivot_table_by_date




```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.


# Sort the resampled pivot table in ascending order on "Total Sales".



# Make sure 'invoice_date' is a datetime dtype
combined_data['invoice_date'] = pd.to_datetime(combined_data['invoice_date'])

# Create the initial pivot table
pivot_table_by_date = combined_data.pivot_table(index='invoice_date', values='total_sales', aggfunc='sum')

# Resample the pivot table into daily bins
daily_sales = pivot_table_by_date.resample('D').sum()

# Rename the 'total_sales' column to 'Total Sales'
daily_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Sort the resampled pivot table in ascending order on "Total Sales"
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=False)

# Display the sorted resampled pivot table
sorted_daily_sales


```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.


# Sort the resampled pivot table in ascending order on "Total Sales".
import pandas as pd

# Assuming your DataFrame is named 'combined_data'
# Convert 'invoice_date' to datetime if it's not already
combined_data['invoice_date'] = pd.to_datetime(combined_data['invoice_date'])

# Create a pivot table with 'invoice_date' as the index and 'total_sales' as the values
pivot_table_by_date = combined_data.pivot_table(index='invoice_date', values='total_sales', aggfunc='sum')

# Resample to weekly bins and sum to get total sales for each week
weekly_sales = pivot_table_by_date.resample('W').sum()

# Rename the 'total_sales' column to 'Total Sales' for clarity
weekly_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Sort the resampled table in ascending order by 'Total Sales'
sorted_weekly_sales = weekly_sales.sort_values('Total Sales', ascending=False)

# Display the sorted weekly sales
sorted_weekly_sales
```
